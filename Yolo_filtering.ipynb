Check file in directory

```python
import os

def count_files_in_subdirectories(directory):
    for root, dirs, files in os.walk(directory):
        # Chỉ quan tâm đến các thư mục con trực tiếp
        if root == directory:
            for subdir in dirs:
                subdir_path = os.path.join(root, subdir)
                file_count = sum([len(files) for _, _, files in os.walk(subdir_path)])
                print(f'Thư mục {subdir} có {file_count} file')

directory_path = 'Data_real'  # Thay thế bằng đường dẫn đến thư mục của bạn
count_files_in_subdirectories(directory_path)
```
